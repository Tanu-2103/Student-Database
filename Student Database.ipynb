```python
n=int(input('Total no of students: '))
name=['']*n
marks=[0]*n
while(1):
    #menu
    print('What would you like to do to the database')
    print('1.Add a student name')
    print('2.Delete a student name')
    print('3.View a student name and marks by roll no.')
    print('4.View a student roll no. and marks by name')
    print('5.Search whether name exist in the database or not')
    print('Any other no. to exit')
    c=int(input())

    #add
    if(c==1):
        while(1):
            print('Enter the roll no. that you want to add')
            r=int(input())
            if(r>n):
                print("There are",n,"no. of students,roll no.",r,"cant be added")
            elif(name[r-1]!=0):
                print('Enter the name that you want to add')
                name[r-1]=input()
                print('Enter the marks that you want to add')
                marks[r-1]=input()
            else:
                print('Roll no.',r,'is already taken by some other student')

            print("Do you want to add another student's id")
            print('1. Yes  2. No')
            k=int(input())
            if(k!=1):
                break
            #for i in range(n):
            #   r=int(input("Enter name,marks,roll no. resp by space: ").split())
            #   name[r[2]]=r[0]\/\/\name[r[-1]]=r[0]
            #   marks[int(r[2])]=r[1]
            
    #delete
    elif(c==2):
        print('Enter the roll no. that you want to delete')
        r=int(input())
        if(name[r-1]==0):
            print("There is no name registered in that roll no.")
        else:
            name[r-1]=0
            marks[r-1]=0

    #view by roll
    elif(c==3):
        print('Enter the roll no. whose details you want to view')
        r=int(input())
        if(name[r-1]==0):
            print("There is no name registered in that roll no.")
        else:
            print("Name",name[r-1],"Marks",marks[r-1])

    #view by name
    elif(c==4):
        print('Enter the name whose marks whose details you want to view')
        s=input()
        r=name.index(s)
        if(r==-1):
            print("There is no name registered in that roll no.")
        else:
            print("Roll No.",r+1,"Marks",marks[r])

    #whether the name exist
    elif(c==5):
        s=input()
        if s in name:
            print("The",s,"exist in the data base")
        else:
            print("The",s,"doesnot exist in the data base")
    else:
        break
```
